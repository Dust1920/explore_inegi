```python
import pandas as pd
import tabula
import os
import camelot
pdf_path = os.path.relpath('fd_enadid09.pdf')

dfs = tabula.read_pdf(pdf_path,pages= 'all')
```

```python
ldfs = dfs.copy()
```

```python
test = ldfs[0].copy()

```

```python
sub_title = list(test.iloc[0, :])
title = list(test.columns)

for i in range(len(title)):
    title[i] = str(title[i]) + " " + str(sub_title[i])
    title[i] = title[i].replace('-','').replace('nan','').rstrip()
title[0] = title[0].replace(' ','')

test.columns = title
test = test.drop(index = [0, 1], axis = 0)
test = test.ffill().reset_index().drop(columns = ['index','LARGO'])
test
```

```python

writer = pd.ExcelWriter('pdf.xlsx',engine='xlsxwriter')
j = 0
for df in ldfs:
    print(j)
    sub_title = list(df.iloc[0, :])
    title = list(df.columns)
    for i in range(len(title)):
        title[i] = str(title[i]) + " " + str(sub_title[i])
        title[i] = title[i].replace('-','').replace('nan','').rstrip()
    title[0] = title[0].replace(' ','')
    df.columns = title
    df = df.drop(index = [0, 1], axis = 0)
    df = df.ffill().reset_index()
    df.to_excel(writer, sheet_name=str(j))
    j += 1
writer.close() 
```
